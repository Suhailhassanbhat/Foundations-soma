```python
print("Clearing out our dataset")
list = []
for page_num in range(1, 16):
    url = f"https://www.epicurious.com/search/cucumbers?page={page_num}"
    print("Now scraping", url)
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)
    
    recipes = doc.find_all("article")

    for recipe in recipes:
        row = {}

        try:
            row['category'] = recipe.find(class_="tag").text
        except:
            pass

        try:
            row['name'] = recipe.find(class_="hed").text
        except:
            pass

        try:
            row['summary'] = recipe.find(class_="dek").text
        except:
            pass

        try:
            row['rating'] = recipe.find(class_="rating").text.split('/',1)[0]
        except:
            pass

        try:
            row['make_again'] = recipe.find(class_="make-again-percentage").text
        except:
            pass
        try:
            row['link'] = recipe.find(class_="view-complete-item")['href']
        except:
            pass
    
        list.append(row)
list
```

```python
df = pd.DataFrame(list)
df = df.dropna(how='all')
df
```

```python
df.to_csv("Epicurious_part1.csv", index=False)
```
