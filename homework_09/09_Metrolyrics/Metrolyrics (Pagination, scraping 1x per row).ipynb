```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
song_list = []
lines = doc.find('tbody').find_all("tr")
for line in lines:
    row = {}
    
    row['song_title'] = line.find("a").text.strip()
    row['URL'] = line.find("a")['href']
    row['year'] = line.find_all("td")[2]['content']
    row['popularity'] = line.find_all('td')[3].find('span')['class'][1].split('r', 1)[1]
    
    song_list.append(row)
print(song_list)
```

```python
all_songs = []

for page_num in range(0, 5):
    url = f"http://www.metrolyrics.com/rem-alpage-{page_num}.html"   
    print("Now scraping", url)

    response = requests.get(url)
    doc = BeautifulSoup(response.text)

    lines = doc.find('tbody').find_all("tr")
    for line in lines:
        row = {}

        row['song_title'] = line.find("a").text.strip()
        row['URL'] = line.find("a")['href']
        row['year'] = line.find_all("td")[2]['content']
        row['popularity'] = line.find_all('td')[3].find('span')['class'][1].split('r', 1)[1]

        all_songs.append(row)

all_songs
    
    
```

```python
df = pd.DataFrame(all_songs)
df
```

```python
df.to_csv("Metrolyrics1.csv", index=False)
```
