```python
def scrape_page(row):
    url = f"{row['URL']}"
    print("Scraping", url)
    
    response = requests.get(url)
    doc = BeautifulSoup(response.text)

    page = {}
    
    page['lyrics'] = [lyric.text for lyric in doc.find('div', attrs={"id" : "lyrics-body-text"}).find_all(class_="verse")]
    
    return pd.Series(page)
```

```python
scrape_page(df.loc[0])
```

```python
scrape_page(df.loc[1])
```

```python
scraped_df = df.apply(scrape_page, axis=1)
scraped_df
```

```python
scraped_df.shape
```

```python
df.head()
```

```python
df = df.merge(scraped_df, left_index=True, right_index=True)
df
```

```python
df.to_csv("Metrolyrics2.csv", index=False)
```
