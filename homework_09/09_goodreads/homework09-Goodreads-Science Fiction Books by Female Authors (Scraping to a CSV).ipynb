```python
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.goodreads.com/list/show/6934.Science_Fiction_Books_by_Female_Authors')
doc = BeautifulSoup(response.text)
```

```python
doc.find(class_="bookTitle").text.split("(",2)[0].strip()
```

```python
doc.find(class_="authorName").text
```

```python
doc.find(class_="minirating").text.split('avg', 2)[0]
```

```python
doc.find(class_="smallText uitext").find('a').text.replace("score:", "")
```

```python
doc.find(class_="number").text.strip()
```

```python
doc.find(class_="tableList js-dataTooltip").find(class_="smallText uitext").find('a').text.strip()
```

```python
book_list = []
books = doc.find_all('div', attrs= {"id":"all_votes"})[0].find_all("tr")
for book in books:
    row = {}
    
    row['Title'] = book.find(class_="bookTitle").text.split("(",2)[0].strip()
    row['Author'] = book.find(class_="authorName").text.strip()
    row['Rating'] = book.find(class_="minirating").text.split('avg', 2)[0]
    row['Rank'] = book.find(class_="number").text.strip()
    row['Score'] = book.find(class_="smallText uitext").find('a').text.replace("score:", "")
    
    
    
    book_list.append(row)
book_list
```

```python
df = pd.DataFrame(book_list)
df
```

```python
df.to_csv("goodreads_100.csv", index=False)
```
