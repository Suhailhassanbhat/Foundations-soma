```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_csv("social-spending.csv")
df['Social Spending'] = df['Social Spending'].str.replace("%", "").astype(float)
df = df.sort_values(by='Social Spending', ascending=True)
df['new_column'] = df['Country'] + " " + df['Social Spending'].astype(str)

def build_colors(row):
    if row['Country'] == 'France':
        return 'blue'
    elif row['Country'] == 'Sweden':
        return 'blue'
    elif row['Country'] == 'Norway':
        return 'blue'
    elif row['Country'] == 'O.E.C.D. avg.':
        return 'blue'
    elif row['Country'] == 'United States':
        return 'blue'
    else:
        return 'lightblue'
colors = df.reset_index().apply(build_colors, axis=1)

df.plot(x='new_column', y='Social Spending', kind='barh', color=colors, width =0.5,figsize =(10,20)).set_title("Public social spending as a share of GDP")
plt.box(False)
plt.savefig("NYT_spending.pdf")
```

```python
df = pd.read_csv("generic_poll_lead.csv")
df = df.sort_values(by='year', ascending=True)

def build_colors(row):
    if row['in_power'] == 'yes':
        return 'lightgrey'
    else:
        return 'yellow'
colors = df.reset_index().apply(build_colors, axis=1)

ax = df.plot(x='year', y='lead', kind='barh', color=colors, figsize=(8,8), fontsize=13).set_title("2018 doesn't look quite like other decision cycles")
plt.box(False)
plt.savefig("NYT_poll.pdf")
```

```python
df = pd.read_csv("cook_pvi.csv")
df['new_column'] = 100 - df['pct_defeated']
colors = ["#802B4E", "#E68CA7"]
ax = df.sort_values(by='pct_defeated', ascending=True).plot(x='cook_pvi', y=['pct_defeated', 'new_column'], color=colors, stacked=True, kind='barh')
plt.legend('')
plt.box(False)
plt.xticks([])
plt.text(x = -15, y = 9, s = 'Incumbents are typically favored to win, even in difficult districts in a wave election',
              fontsize = 14, weight = 'bold', alpha = .75)
plt.text(x = -15, y = 8.5,
              s = 'Incumbents Are Tough to Beat, Even in a Wave',
             fontsize = 12, alpha = .85)
plt.text(x = -15, y = 8,
              s = 'Winning percentage for incumbents representing the presidents party in 2006 and 2010, by their partys advantage in Cook P.V.I.',
             fontsize = 10, alpha = .85)
plt.text(x = 20, y = 7.6,
              s = 'Incumbent Defeated.',
             fontsize = 10, alpha = .85)
plt.text(x = 70, y = 7.6,
              s = 'Incumbent Survives',
             fontsize = 10, alpha = .85)

plt.savefig("NYT_cook_pvi.pdf")
```
