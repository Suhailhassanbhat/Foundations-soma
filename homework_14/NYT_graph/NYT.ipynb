```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['pdf.fonttype'] = 42
```

```python
df = pd.read_csv("social-spending.csv")
df['Social Spending'] = df['Social Spending'].str.replace("%", "").astype(float)
df = df.sort_values(by='Social Spending', ascending=True)
```

```python
df['new_column'] = df['Country'] + " " + df['Social Spending'].astype(str)
```

```python
def build_colors(row):
    if row['Country'] == 'France':
        return 'blue'
    elif row['Country'] == 'Sweden':
        return 'blue'
    elif row['Country'] == 'Norway':
        return 'blue'
    elif row['Country'] == 'O.E.C.D. avg.':
        return 'blue'
    elif row['Country'] == 'United States':
        return 'blue'
    else:
        return 'lightblue'
colors = df.reset_index().apply(build_colors, axis=1)
```

```python
df.plot(x='new_column', y='Social Spending', kind='barh', color=colors, width =0.8,figsize =(10,20)).set_title("Public social spending as a share of GDP")
plt.savefig("NYT_spending.pdf")
```

```python
df = pd.read_csv("generic_poll_lead.csv")
df = df.sort_values(by='year', ascending=True)
```

```python
def build_colors(row):
    if row['in_power'] == 'yes':
        return 'lightgrey'
    else:
        return 'yellow'
colors = df.reset_index().apply(build_colors, axis=1)
```

```python
ax = df.plot(x='year', y='lead', kind='barh', color=colors, figsize=(8,8), fontsize=13).set_title("2018 doesn't look quite like other decision cycles")
plt.savefig("NYT_polling.pdf")
```

```python
df = pd.read_csv("cook_pvi.csv")
df['new_column'] = 100 - df['pct_defeated']
```

```python
ax = df.sort_values(by='pct_defeated', ascending=True).plot(x='cook_pvi', y=['pct_defeated', 'new_column'], stacked=True, kind='barh')
ax.set_title("Incumbents are tough to beat. Even in a wave", fontsize=14)
plt.savefig("NYT_cook_pvi.pdf")
```
